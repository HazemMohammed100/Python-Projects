```python
import pandas as pd
df_1 = pd.read_csv('Names_1.csv')
df_2 = pd.read_csv('Names_2.csv')
```

```python
df_1.info()
df_1
```

```python
df_2.info()
df_2
```

# Using The Join Method

### `1. ` Inner Join

```python
df_inner = (df_1.set_index('name')).join(df_2.set_index('name'), on=['name'], how='inner', lsuffix='_1', rsuffix='_2')
df_inner.reset_index(inplace=True)
df_inner
```

### `2. ` Left Join

```python
df_left = (df_1.set_index('name')).join(df_2.set_index('name'), on=['name'], how='left', lsuffix='_1', rsuffix='_2')
df_left.reset_index(inplace=True)
df_left
```

### `3. ` Right Join

```python
df_right = (df_1.set_index('name')).join(df_2.set_index('name'), on=['name'], how='right', lsuffix='_1', rsuffix='_2')
df_right.reset_index(inplace=True, drop=True)
df_right
```

# Using The Merge Method

### `1. ` Inner Join

```python
df_inner = pd.merge(df_1, df_2, on=['name'], how='inner', suffixes=('_1','_2'))
df_inner
```

### `2. ` Left Join

```python
df_left = pd.merge(df_1, df_2, on=['name'], how='left', suffixes=('_1','_2'))
df_left
```

### `3. ` Right Join

```python
df_right = pd.merge(df_1, df_2, on=['name'], how='right', suffixes=('_1','_2'))
df_right
```

# Conclusion:

- **Inner Join:** produces only the rows that are available in both dataframes based on key (column/columns) specified.
<br>

- **Left Join:** produces all the rows from the first dataframe and matches the rows from the second dataframe (if available) based on key the column/columns specified. If there is no match, the right side will contain null.
<br>

- **Right Join:** produces all the rows from the second dataframe and matches the rows from the first dataframe (if available) based on key the column/columns specified. If there is no match, the left side will contain null.
